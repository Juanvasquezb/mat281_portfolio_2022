EL conjunto de datos se denomina `company_sales_data.csv`, el cual contiene información tal como: número del mes, unidades, precio, etc.

Lo primero es cargar el conjunto de datos y ver las primeras filas que lo componen:

```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(10,8)})
```

```python
# cargar datos
df = pd.read_csv("https://raw.githubusercontent.com/fralfaro/MAT281_2022/main/labs/lab_04/data/company_sales_data.csv")
df.head()
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos. Para cumplir este objetivo debe resolver las siguientes problemáticas:

**Observación.-** Puedes ocupar las librerías de Matplolib o Seaborn.

1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
month_number = df['month_number']
total_profit = df['total_profit']

fig = plt.figure(figsize=(10, 5)) 
plt.plot(month_number, total_profit,color="black",linewidth=1)
plt.title("Gráfico Lineal") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('total_profit') # nombre eje y
plt.show()

fig = plt.figure(figsize=(10, 5)) 
plt.scatter(month_number, total_profit,color="black",linewidth=1)
plt.title("Gráfico de Dispersión") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('total_profit') # nombre eje y
plt.show()
```

Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
# Supuesto: Mostrar en un gráfico todos los puntos que se venden por mes
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

fig = plt.figure(figsize=(10, 5)) 
plt.plot(month_number,facecream,label='facecream',color="black",linewidth=1)
plt.plot(month_number,facewash,label='facewash',color="yellow", linewidth=1)
plt.plot(month_number,toothpaste,label='toothpaste',color="blue",linewidth=1)
plt.plot(month_number,bathingsoap,label='bathingsoap',color="red",linewidth=1)
plt.plot(month_number,shampoo,label='shampoo',color="magenta",linewidth=1)
plt.plot(month_number,moisturizer,label='moisturizer',color="green",linestyle = "--",linewidth=1)
plt.legend() 
plt.title("Gráfico de Venta de Productos") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('Producto') # nombre eje y
plt.show()
```

3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
fig = plt.figure(figsize=(10, 5))
width = 0.25
plt.bar(month_number - width/2,facecream,width=width,color="black",alpha=0.6, label = 'facecream')
plt.bar(month_number + width/2,facewash,width=width,color="red",alpha=0.6, label='facewash')
plt.legend() 
plt.title("Gráfico de Venta de Productos") # agregar titulo
plt.xlabel('month_number') # nombre eje x
plt.ylabel('Cantidad de Producto') # nombre eje y
plt.show()
```

4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
productos = df.drop(['month_number','total_units','total_profit'],axis=1) # elimina variables que no se utilizan
plt.figure(figsize=(10, 6))
sns.boxplot(data=productos)
plt.title("Gráfico de Cajas") # agregar titulo
plt.xlabel('Producto') # nombre eje x
plt.ylabel('Cantidad') # nombre eje y
plt.show()
```

5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
facecream = sum(df['facecream'])
facewash = sum(df['facewash'])
toothpaste = sum(df['toothpaste'])
bathingsoap = sum(df['bathingsoap'])
shampoo = sum(df['shampoo'])
moisturizer = sum(df['moisturizer'])
total = sum(df["total_units"])

valores = [facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
nombres = ["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"]
plt.pie(valores, labels=nombres, autopct="%0.1f %%")
plt.axis("equal")
plt.show()

"El total de datos es de: "+ str(total)
```
